## Copy a subset of .tif files from a massive unsorted folder into individual folders per sample, as determined by the R##### in part of the file name

```python
import shutil
import glob
import os
from pathlib import Path
```

Make sure to change the target folder. It should be the folder with all of the .tif MIPs in it.

```python
targetFolder = "G:/20220518_160557_051822h2bgfp6hpfish Mips/"
fileList = glob.glob(targetFolder+'S000_t000000_V000_R*.tif')
```

```python
#just to check, should show the files you want to copy
fileList
```

```python
for file in fileList:
    file = Path(file)
    name = file.stem
    current_position = name.split('_')[3]
    outputDirectory = targetFolder+"/"+current_position
    outputDirectoryExistsCheck = os.path.exists(outputDirectory)
    if not outputDirectoryExistsCheck:
        os.mkdir(outputDirectory)
    files_to_move = glob.glob(targetFolder+'*'+current_position+'*.*')
    #print(files_to_move)
    for temp in files_to_move:
        time_check = Path(temp).stem.split('_')[1][1:]
        #print(time_check)
        if (int(time_check)%4==0):
            shutil.copyfile(Path(temp), outputDirectory+"/"+Path(temp).stem+Path(temp).suffix)

        #print(Path(temp).stem+Path(temp).suffix)
```
